```python
from data import * 
from lda_trial import *

from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
```

```python
train_pos = IMDBDataset('train', data_limit=20_000, load_class='pos')
train_neg = IMDBDataset('train', data_limit=20_000, load_class='neg')

trainset = IMDBDataset('train', data_limit=20_000, load_class='both')
validset = IMDBDataset('valid')
```

```python
n = 20000

for k in [2, 3, 5, 10, 20, 50, 100]:
    print(k)
    pos_model, final_metric = tp_one_trial(train_pos, 'lda', k, n, 
                                                  3, 5,  # args.burn_in,
                                                  max_iter=20, stop_increase=5, metric='ll')

    neg_model, final_metric = tp_one_trial(train_neg, 'lda', k, n, 
                                                  3, 5,  # args.burn_in,
                                                  max_iter=20, stop_increase=5, metric='ll')

    lda_pos, lda_y = load_LDA_data_batch(pos_model, trainset)
    lda_neg, lda_y = load_LDA_data_batch(neg_model, trainset)
    lda_x = np.concatenate([lda_pos, lda_neg], axis=1)

    model = LinearSVC()
    model.fit(lda_x, lda_y)

    prediction = model.predict(lda_x)
    ground_truth = lda_y
    print(f'Train: {100*accuracy_score(prediction, ground_truth):6.5f}%')

    lda_pos_valid, lda_y_valid = load_LDA_data_batch(pos_model, validset)
    lda_neg_valid, lda_y_valid = load_LDA_data_batch(neg_model, validset)
    lda_valid = np.concatenate([lda_pos_valid, lda_neg_valid], axis=1)

    prediction = model.predict(lda_valid)
    ground_truth = lda_y_valid
    print(f'Test: {100*accuracy_score(prediction, ground_truth):6.5f}%')
    print('-' * 100)
```
